```
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris["data"],
                  columns=iris["feature_names"])
df["answer"]=iris["target"]
df.to_csv("iris.csv",encoding="utf-8", index=False) #index是去除標號
df 
```

```
# list(x), pandas(x)->用numpy
import numpy as np
from sklearn.model_selection import train_test_split
x = df.drop(["answer"], axis=1)
y = df["answer"]
x, y = np.array(x), np.array(y)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1) #135/15
```

# 決策樹 

```
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=2)
clf.fit(x_train, y_train)
```

```
import matplotlib.pyplot as plt #matplot python畫圖工具 
from sklearn.tree import plot_tree
plt.figure(figsize=(30, 30)) #figsize控制下載大小 10英吋 寬高
tree = plot_tree(clf, feature_names=iris["feature_names"], #給上名字
                 class_names=iris["target_names"],  #最多的給上名字
                 filled=True)                       #塗色
plt.show
```

```
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test)
accuracy_score(pre, y_test)
```
